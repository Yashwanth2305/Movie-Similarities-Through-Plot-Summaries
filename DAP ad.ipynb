```python
from nltk.stem.snowball import SnowballStemmer
```

```python
stemmer = SnowballStemmer("english")
```

```python
#a function that process the texts in the plots using stop words and also for normailizing the text
def text_process(plot):
    no_punc = [char for char in plot if char not in string.punctuation]
    no_punc = ''.join(no_punc)
    no_punc_mod = [word for word in no_punc.split() if word.lower not in stopwords.words('english')]
    stems = [stemmer.stem(word) for word in no_punc_mod]
    return stems
```

```python
#applying the function to the first five rows of the plot column to see
df['plots'].head(5).apply(text_process)
```

```python
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
plots = [
    "Casablanca\n[u' Drama', u' Romance', u' War']",
    "The Godfather\n[u' Crime', u' Drama']",
    "The Shawshank Redemption\n[u' Crime', u' Drama']",
    "Raging Bull[u' Biography', u' Drama', u' Sport']",
    "Titanic[u' Drama', u' Romance']",
    "Rocky[u' Drama', u' Sport']",
    "Patton[u' Biography', u' Drama', u' War']",
    "The Pianist[u' Biography', u' Drama', u' War']",
    "Doctor Zhivago[u' Drama', u' Romance', u' War']",
    "Giant[u' Drama', u' Romance']",
    "Shane[u' Drama', u' Romance', u' Western']"
]
```

```python
tfidf_vectorizer = TfidfVectorizer()
tfidf_matrix = tfidf_vectorizer.fit_transform(plots)

cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)
similar_index = plots.index("Titanic[u' Drama', u' Romance']")
print("Index of movie:", similar_index)
```

```python
def find_similar_movies(movie_index, similarity_matrix, movie_titles, top_n=5):
    similar_indices = np.argsort(similarity_matrix[movie_index])[::-1][1:top_n+1]
    similar_movies = [(movie_titles[i], similarity_matrix[movie_index][i]) for i in similar_indices]
    return similar_movies
similar_movies = find_similar_movies(similar_index, cosine_sim, plots, top_n=1)
most_similar_movie, similarity_score = similar_movies[0]
print(f"The movie most similar to 'Titanic' is '{most_similar_movie}' with a similarity score of {similarity_score:.2f}")
```

```python
import pickle
from sklearn.neighbors import NearestNeighbors
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorizer = TfidfVectorizer(stop_words='english')
X = vectorizer.fit_transform(plots)

# Train a NearestNeighbors model
model = NearestNeighbors(n_neighbors=3, metric='cosine').fit(X)

# Save the model and vectorizer to a pickle file
with open('movie_recommendation_model.pkl', 'wb') as model_file:
    pickle.dump((vectorizer, model, df), model_file)
```
